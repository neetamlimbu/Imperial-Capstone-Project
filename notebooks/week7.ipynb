```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor

# ---------------------------------------------------------
# HARD-CODED WEEKLY INPUTS (WEEKS 1–6)
# ---------------------------------------------------------

week1_inputs = [
    [0.333333, 0.666667],
    [0.777778, 0.222222],
    [0.142857, 0.571429, 0.857143],
    [0.285714, 0.714286, 0.428571, 0.857143],
    [0.0625, 0.5, 0.9375, 0.25],
    [0.111111, 0.444444, 0.777778, 0.222222, 0.888889],
    [0.090909, 0.363636, 0.636364, 0.181818, 0.545455, 0.818182],
    [0.125, 0.25, 0.375, 0.5, 0.625, 0.75, 0.875, 0.0625]
]

week2_inputs = [
    [0.5, 0.5],
    [0.7, 0.3],
    [0.2, 0.6, 0.8],
    [0.2, 0.8, 0.3, 0.7],
    [0.08, 0.52, 0.92, 0.27],
    [0.2, 0.5, 0.8, 0.3, 0.9],
    [0.12, 0.38, 0.66, 0.22, 0.58, 0.84],
    [0.15, 0.275, 0.4, 0.525, 0.65, 0.775, 0.9, 0.1]
]

week3_inputs = [
    [0.45, 0.55],
    [0.725, 0.275],
    [0.8, 0.2, 0.4],
    [0.25, 0.75, 0.35, 0.65],
    [0.07, 0.51, 0.93, 0.26],
    [0.21, 0.49, 0.81, 0.31, 0.91],
    [0.1, 0.36, 0.64, 0.2, 0.56, 0.82],
    [0.13, 0.26, 0.38, 0.51, 0.63, 0.76, 0.88, 0.07]
]

week4_inputs = [
    [0.427778, 0.572222],
    [0.734259, 0.265741],
    [0.380952, 0.457143, 0.685714],
    [0.245238, 0.754762, 0.359524, 0.735714],
    [0.070833, 0.51, 0.929167, 0.26],
    [0.173704, 0.478148, 0.795926, 0.277407, 0.89963],
    [0.103636, 0.367879, 0.645455, 0.200606, 0.561818, 0.826061],
    [0.135, 0.261667, 0.385, 0.511667, 0.635, 0.761667, 0.885, 0.0775]
]

week5_inputs = [
    [0.42, 0.58],
    [0.63, 0.37],
    [0.31, 0.52, 0.74],
    [0.27, 0.61, 0.44, 0.83],
    [0.18, 0.57, 0.89, 0.33],
    [0.22, 0.48, 0.79, 0.29, 0.91],
    [0.14, 0.39, 0.66, 0.23, 0.58, 0.84],
    [0.16, 0.28, 0.41, 0.53, 0.66, 0.78, 0.9, 0.11]
]

week6_inputs = [
    [0.436555, 0.561852],
    [0.643711, 0.350427],
    [0.290003, 0.526542, 0.748973],
    [0.257349, 0.623142, 0.4314, 0.843698],
    [0.164072, 0.559418, 0.90988, 0.318913],
    [0.203885, 0.477786, 0.796998, 0.271818, 0.890277],
    [0.133556, 0.371272, 0.667798, 0.244733, 0.565004, 0.824401],
    [0.140764, 0.29475, 0.39968, 0.531209, 0.640643, 0.763956, 0.880412, 0.096299]
]

# ---------------------------------------------------------
# HARD-CODED WEEKLY OUTPUTS (WEEKS 1–6)
# ---------------------------------------------------------

week1_outputs = [
    5.724814756293084e-48,
    0.1668524240531267,
    -0.03513300955356602,
    -16.181877221146838,
    94.62121154785156,
    -1.7666841443101928,
    1.062660972512584,
    8.668221875
]

week2_outputs = [
    2.6752879910742468e-09,
    0.43801389905941557,
    -0.06505619615899967,
    -15.300482957509526,
    73.84870222928335,
    -1.721714026598003,
    0.8380519063103001,
    8.528175
]

week3_outputs = [
    1.5542034905615116e-13,
    0.41155118475915997,
    -0.038979703371155035,
    -11.860925282389413,
    85.55341253928546,
    -1.8218346425529346,
    1.0054527181773487,
    8.63406
]

week4_outputs = [
    1.4138862577600168e-16,
    0.3110015499941023,
    -0.11851799494062873,
    -14.10998365700517,
    84.26606790996709,
    -1.8336621537886206,
    0.9691878328431802,
    8.611369718333
]

week5_outputs = [
    3.4004559045423516e-18,
    0.3658481167000386,
    -0.11876831843175635,
    -13.16435158471511,
    54.36741321156047,
    -1.7568980194796602,
    0.8328963722243514,
    8.50409
]

week6_outputs = [
    9.045368553687896e-15,
    0.40670960313561644,
    -0.10857204651762475,
    -13.893143435467305,
    74.42395745271067,
    -1.8021181616140751,
    0.9452093352908661,
    8.5785592753384
]

all_inputs = [week1_inputs, week2_inputs, week3_inputs, week4_inputs, week5_inputs, week6_inputs]
all_outputs = [week1_outputs, week2_outputs, week3_outputs, week4_outputs, week5_outputs, week6_outputs]

# ---------------------------------------------------------
# MODEL ZOO (WITH MULTIPLE NEURAL NETWORKS)
# ---------------------------------------------------------

def get_models():
    return [
        # Tree-based models
        RandomForestRegressor(
            n_estimators=300,
            max_depth=None,
            random_state=0
        ),
        GradientBoostingRegressor(
            n_estimators=400,
            learning_rate=0.05,
            max_depth=3,
            random_state=0
        ),

        # Kernel model
        SVR(kernel='rbf', C=20, gamma='scale'),

        # Neural Network 1 — medium, ReLU
        MLPRegressor(
            hidden_layer_sizes=(64, 64),
            activation='relu',
            solver='adam',
            learning_rate_init=0.001,
            max_iter=5000,
            random_state=0
        ),

        # Neural Network 2 — deeper, tanh
        MLPRegressor(
            hidden_layer_sizes=(128, 128, 64),
            activation='tanh',
            solver='adam',
            learning_rate_init=0.0007,
            max_iter=7000,
            random_state=1
        ),

        # Neural Network 3 — wide, ReLU
        MLPRegressor(
            hidden_layer_sizes=(256, 128),
            activation='relu',
            solver='adam',
            learning_rate_init=0.0005,
            max_iter=7000,
            random_state=2
        )
    ]

def ensemble_predict(models, X_query):
    preds = np.column_stack([m.predict(X_query) for m in models])
    return preds.mean(axis=1)

# ---------------------------------------------------------
# BUILD PER-FUNCTION DATASETS
# ---------------------------------------------------------

def build_function_data():
    X_f = {}
    y_f = {}
    for f in range(8):
        X_list = [np.array(w[f], dtype=float) for w in all_inputs]
        X = np.vstack(X_list)  # shape (6, d_f)
        y = np.array([w[f] for w in all_outputs], dtype=float)
        X_f[f + 1] = X
        y_f[f + 1] = y
    return X_f, y_f

# ---------------------------------------------------------
# PROPOSE WEEK 7 OPTIMUM POINTS (RANDOM SEARCH + ENSEMBLE)
# ---------------------------------------------------------

def propose_week7_points(maximise=True, n_candidates=20000, lower=0.0, upper=1.0, seed=42):
    X_f, y_f = build_function_data()
    rng = np.random.RandomState(seed)
    week7_inputs = []

    for f_id in range(1, 9):
        X = X_f[f_id]
        y = y_f[f_id]
        d = X.shape[1]

        models = get_models()
        for m in models:
            m.fit(X, y)

        candidates = rng.uniform(lower, upper, size=(n_candidates, d))
        y_pred = ensemble_predict(models, candidates)

        idx_best = np.argmax(y_pred) if maximise else np.argmin(y_pred)
        x_best = candidates[idx_best]
        week7_inputs.append(x_best.tolist())

    return week7_inputs

# ---------------------------------------------------------
# UTILITY: PORTAL FORMAT
# ---------------------------------------------------------

def format_for_portal(x):
    return " - ".join(f"{v:.6f}" for v in x)

# ---------------------------------------------------------
# RUN: FIND WEEK 7 OPTIMUM POINTS
# ---------------------------------------------------------

if __name__ == "__main__":
    # Set maximise=False if a function is minimisation; here we assume maximise.
    week7_inputs = propose_week7_points(maximise=True)

    print("=== WEEK 7 PROPOSED INPUTS (OPTIMUM ESTIMATES) ===")
    for i, w in enumerate(week7_inputs, start=1):
        print(f"Function {i}: {format_for_portal(w)}")
```
