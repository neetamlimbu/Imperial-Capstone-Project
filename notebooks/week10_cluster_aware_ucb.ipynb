```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor

# =========================================================
# 1. DATA: WEEK 1–9 INPUTS AND OUTPUTS
#    (You can extend with week10_inputs/week10_outputs later)
# =========================================================

week1_inputs = [
    [0.333333, 0.666667],
    [0.777778, 0.222222],
    [0.142857, 0.571429, 0.857143],
    [0.285714, 0.714286, 0.428571, 0.857143],
    [0.0625, 0.5, 0.9375, 0.25],
    [0.111111, 0.444444, 0.777778, 0.222222, 0.888889],
    [0.090909, 0.363636, 0.636364, 0.181818, 0.545455, 0.818182],
    [0.125, 0.25, 0.375, 0.5, 0.625, 0.75, 0.875, 0.0625]
]

week2_inputs = [
    [0.5, 0.5],
    [0.7, 0.3],
    [0.2, 0.6, 0.8],
    [0.2, 0.8, 0.3, 0.7],
    [0.08, 0.52, 0.92, 0.27],
    [0.2, 0.5, 0.8, 0.3, 0.9],
    [0.12, 0.38, 0.66, 0.22, 0.58, 0.84],
    [0.15, 0.275, 0.4, 0.525, 0.65, 0.775, 0.9, 0.1]
]

week3_inputs = [
    [0.45, 0.55],
    [0.725, 0.275],
    [0.8, 0.2, 0.4],
    [0.25, 0.75, 0.35, 0.65],
    [0.07, 0.51, 0.93, 0.26],
    [0.21, 0.49, 0.81, 0.31, 0.91],
    [0.1, 0.36, 0.64, 0.2, 0.56, 0.82],
    [0.13, 0.26, 0.38, 0.51, 0.63, 0.76, 0.88, 0.07]
]

week4_inputs = [
    [0.427778, 0.572222],
    [0.734259, 0.265741],
    [0.380952, 0.457143, 0.685714],
    [0.245238, 0.754762, 0.359524, 0.735714],
    [0.070833, 0.51, 0.929167, 0.26],
    [0.173704, 0.478148, 0.795926, 0.277407, 0.89963],
    [0.103636, 0.367879, 0.645455, 0.200606, 0.561818, 0.826061],
    [0.135, 0.261667, 0.385, 0.511667, 0.635, 0.761667, 0.885, 0.0775]
]

week5_inputs = [
    [0.42, 0.58],
    [0.63, 0.37],
    [0.31, 0.52, 0.74],
    [0.27, 0.61, 0.44, 0.83],
    [0.18, 0.57, 0.89, 0.33],
    [0.22, 0.48, 0.79, 0.29, 0.91],
    [0.14, 0.39, 0.66, 0.23, 0.58, 0.84],
    [0.16, 0.28, 0.41, 0.53, 0.66, 0.78, 0.9, 0.11]
]

week6_inputs = [
    [0.436555, 0.561852],
    [0.643711, 0.350427],
    [0.290003, 0.526542, 0.748973],
    [0.257349, 0.623142, 0.4314, 0.843698],
    [0.164072, 0.559418, 0.90988, 0.318913],
    [0.203885, 0.477786, 0.796998, 0.271818, 0.890277],
    [0.133556, 0.371272, 0.667798, 0.244733, 0.565004, 0.824401],
    [0.140764, 0.29475, 0.39968, 0.531209, 0.640643, 0.763956, 0.880412, 0.096299]
]

week7_inputs = [
    [0.999509, 0.605025],
    [0.709894, 0.999867],
    [0.018005, 0.040702, 0.02947],
    [0.022476, 0.05575, 0.00098, 0.032298],
    [0.990736, 0.060822, 0.932835, 0.009122],
    [0.565687, 0.022461, 0.000202, 0.168316, 0.021242],
    [0.998542, 0.97969, 0.934707, 0.047897, 0.948514, 0.802608],
    [0.731876, 0.927411, 0.97031, 0.985745, 0.915076, 0.597028, 0.907676, 0.859568]
]

week8_inputs = [
    [0.997175, 0.87883],
    [0.709894, 0.999867],
    [0.019391, 0.963731, 0.833755],
    [0.26927, 0.607255, 0.583638, 0.528609],
    [0.990736, 0.060822, 0.932835, 0.009122],
    [0.117249, 0.024704, 0.004459, 0.090832, 0.076921],
    [0.016213, 0.111646, 0.428254, 0.681926, 0.523978, 0.977388],
    [0.283187, 0.111804, 0.590719, 0.550574, 0.78482, 0.919104, 0.881617, 0.049933]
]

week9_inputs = [
    [0.990198, 0.81937],
    [0.709894, 0.999867],
    [0.037714, 0.991662, 0.857859],
    [0.264095, 0.59031, 0.575603, 0.333135],
    [0.990736, 0.060822, 0.932835, 0.009122],
    [0.051082, 0.041178, 0.03159, 0.136823, 0.343311],
    [0.059048, 0.366885, 0.960575, 0.052238, 0.469479, 0.879831],
    [0.049164, 0.319656, 0.074848, 0.58412, 0.404708, 0.887928, 0.792066, 0.003256]
]

week1_outputs = [
    5.724814756293084e-48, 0.1668524240531267, -0.03513300955356602,
    -16.181877221146838, 94.62121154785156, -1.7666841443101928,
    1.062660972512584, 8.668221875
]

week2_outputs = [
    2.6752879910742468e-09, 0.43801389905941557, -0.06505619615899967,
    -15.300482957509526, 73.84870222928335, -1.721714026598003,
    0.8380519063103001, 8.528175
]

week3_outputs = [
    1.5542034905615116e-13, 0.41155118475915997, -0.038979703371155035,
    -11.860925282389413, 85.55341253928546, -1.8218346425529346,
    1.0054527181773487, 8.63406
]

week4_outputs = [
    1.4138862577600168e-16, 0.3110015499941023, -0.11851799494062873,
    -14.10998365700517, 84.26606790996709, -1.8336621537886206,
    0.9691878328431802, 8.611369718333
]

week5_outputs = [
    3.4004559045423516e-18, 0.3658481167000386, -0.11876831843175635,
    -13.16435158471511, 54.36741321156047, -1.7568980194796602,
    0.8328963722243514, 8.50409
]

week6_outputs = [
    9.045368553687896e-15, 0.40670960313561644, -0.10857204651762475,
    -13.893143435467305, 74.42395745271067, -1.8021181616140751,
    0.9452093352908661, 8.5785592753384
]

week7_outputs = [
    -7.415294019870151e-97, 0.5846280632249955, -0.14575467861190483,
    -23.13006014797806, 1084.1093856854866, -1.7200118813875427,
    8.711278013392023e-05, 4.755881312715599
]

week8_outputs = [
    1.0490301671774682e-138, 0.519729654344556, -0.05817503127572031,
    -7.267670167442148, 1084.1093856854866, -1.9940951131996647,
    0.15208551171349755, 7.9989563724441
]

week9_outputs = [
    -9.112272518922075e-118, 0.6166877176848666, -0.07268195557278455,
    -6.26532514156677, 1084.1093856854866, -2.2079469432780883,
    0.567715439829087, 8.8031521874784
]

all_inputs = [
    week1_inputs, week2_inputs, week3_inputs,
    week4_inputs, week5_inputs, week6_inputs,
    week7_inputs, week8_inputs, week9_inputs
]

all_outputs = [
    week1_outputs, week2_outputs, week3_outputs,
    week4_outputs, week5_outputs, week6_outputs,
    week7_outputs, week8_outputs, week9_outputs
]

# =========================================================
# 2. MODEL ZOO + ENSEMBLE
# =========================================================

def get_models():
    return [
        RandomForestRegressor(n_estimators=300, random_state=0),
        GradientBoostingRegressor(n_estimators=400, learning_rate=0.05, max_depth=3, random_state=0),
        SVR(kernel='rbf', C=20, gamma='scale'),
        MLPRegressor(hidden_layer_sizes=(64, 64), activation='relu', max_iter=5000, random_state=0),
        MLPRegressor(hidden_layer_sizes=(128, 128, 64), activation='tanh', max_iter=7000, random_state=1),
        MLPRegressor(hidden_layer_sizes=(256, 128), activation='relu', max_iter=7000, random_state=2)
    ]

def ensemble_predict_full(models, X_query):
    """
    Returns mean and std across ensemble predictions.
    """
    preds = np.column_stack([m.predict(X_query) for m in models])
    mean = preds.mean(axis=1)
    std = preds.std(axis=1)
    return mean, std

# =========================================================
# 3. DATASET BUILDING + WEEK APPEND
# =========================================================

def build_function_data(all_inputs, all_outputs):
    X_f, y_f = {}, {}
    for f in range(8):
        X = np.vstack([np.array(w[f], dtype=float) for w in all_inputs])
        y = np.array([w[f] for w in all_outputs], dtype=float)
        X_f[f + 1] = X
        y_f[f + 1] = y
    return X_f, y_f

def add_new_week(all_inputs, all_outputs, new_inputs, new_outputs):
    all_inputs.append(new_inputs)
    all_outputs.append(new_outputs)
    return all_inputs, all_outputs

# =========================================================
# 4. CANDIDATE GENERATION (GLOBAL + CLUSTER-AWARE)
# =========================================================

def sample_global_candidates(rng, d, n_global=10000):
    return rng.uniform(0.0, 1.0, size=(n_global, d))

def sample_local_candidates(rng, X_hist, y_hist, n_best=3, n_local_per_best=3000, radius=0.05):
    """
    Cluster-aware: pick top n_best historical points, sample around them.
    """
    # sort by y descending (maximisation)
    idx_sorted = np.argsort(-y_hist)
    best_idx = idx_sorted[:n_best]
    best_points = X_hist[best_idx]

    locals_list = []
    for bp in best_points:
        noise = rng.normal(loc=0.0, scale=radius, size=(n_local_per_best, X_hist.shape[1]))
        cand = bp + noise
        cand = np.clip(cand, 0.0, 1.0)
        locals_list.append(cand)

    return np.vstack(locals_list)

def build_candidates(rng, X_hist, y_hist, n_global=8000, n_best=3, n_local_per_best=2000, radius=0.05):
    d = X_hist.shape[1]
    global_cands = sample_global_candidates(rng, d, n_global=n_global)
    local_cands = sample_local_candidates(rng, X_hist, y_hist, n_best=n_best,
                                          n_local_per_best=n_local_per_best, radius=radius)
    return np.vstack([global_cands, local_cands])

# =========================================================
# 5. ACQUISITION FUNCTION
# =========================================================

def acquisition(mean, std, maximise=True, lambda_unc=0.5):
    """
    Simple UCB-style acquisition: mean + lambda * std (for maximise),
    or mean - lambda * std (for minimise).
    """
    if maximise:
        return mean + lambda_unc * std
    else:
        return mean - lambda_unc * std

# =========================================================
# 6. PROPOSE NEXT WEEK POINTS (GENERAL)
# =========================================================

def propose_next_week(all_inputs, all_outputs,
                      maximise=True,
                      n_global=8000,
                      n_best=3,
                      n_local_per_best=2000,
                      radius=0.05,
                      lambda_unc=0.5,
                      seed=42):
    """
    General optimiser:
    - Uses all historical weeks in all_inputs/all_outputs
    - Builds per-function datasets
    - Trains ensemble
    - Samples global + local (cluster-aware) candidates
    - Uses UCB-style acquisition to pick next point
    """
    X_f, y_f = build_function_data(all_inputs, all_outputs)
    rng = np.random.RandomState(seed)
    next_inputs = []

    for f_id in range(1, 9):
        X_hist, y_hist = X_f[f_id], y_f[f_id]
        d = X_hist.shape[1]

        models = get_models()
        for m in models:
            m.fit(X_hist, y_hist)

        candidates = build_candidates(
            rng, X_hist, y_hist,
            n_global=n_global,
            n_best=n_best,
            n_local_per_best=n_local_per_best,
            radius=radius
        )

        mean, std = ensemble_predict_full(models, candidates)
        acq = acquisition(mean, std, maximise=maximise, lambda_unc=lambda_unc)

        idx = np.argmax(acq) if maximise else np.argmin(acq)
        next_inputs.append(candidates[idx].tolist())

    return next_inputs

# =========================================================
# 7. PORTAL FORMAT
# =========================================================

def format_for_portal(x):
    return " - ".join(f"{v:.6f}" for v in x)

# =========================================================
# 8. RUN: GENERATE WEEK 10 POINTS
# =========================================================

if __name__ == "__main__":
    # For now, all_inputs/all_outputs contain Weeks 1–9.
    # This call proposes Week 10 points.
    week10_inputs = propose_next_week(
        all_inputs,
        all_outputs,
        maximise=True,
        n_global=8000,
        n_best=3,
        n_local_per_best=2000,
        radius=0.05,
        lambda_unc=0.5,
        seed=42
    )

    print("=== WEEK 10 PROPOSED INPUTS (CLUSTER-AWARE + UCB) ===")
    for i, w in enumerate(week10_inputs, start=1):
        print(f"Function {i}: {format_for_portal(w)}")
```
