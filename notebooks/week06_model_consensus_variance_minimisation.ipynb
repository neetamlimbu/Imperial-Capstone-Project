```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor

# ---------------------------------------------------------
# HARD-CODED WEEKLY INPUTS
# ---------------------------------------------------------

week1_inputs = [
    [0.333333, 0.666667],
    [0.777778, 0.222222],
    [0.142857, 0.571429, 0.857143],
    [0.285714, 0.714286, 0.428571, 0.857143],
    [0.0625, 0.5, 0.9375, 0.25],
    [0.111111, 0.444444, 0.777778, 0.222222, 0.888889],
    [0.090909, 0.363636, 0.636364, 0.181818, 0.545455, 0.818182],
    [0.125, 0.25, 0.375, 0.5, 0.625, 0.75, 0.875, 0.0625]
]

week2_inputs = [
    [0.5, 0.5],
    [0.7, 0.3],
    [0.2, 0.6, 0.8],
    [0.2, 0.8, 0.3, 0.7],
    [0.08, 0.52, 0.92, 0.27],
    [0.2, 0.5, 0.8, 0.3, 0.9],
    [0.12, 0.38, 0.66, 0.22, 0.58, 0.84],
    [0.15, 0.275, 0.4, 0.525, 0.65, 0.775, 0.9, 0.1]
]

week3_inputs = [
    [0.45, 0.55],
    [0.725, 0.275],
    [0.8, 0.2, 0.4],
    [0.25, 0.75, 0.35, 0.65],
    [0.07, 0.51, 0.93, 0.26],
    [0.21, 0.49, 0.81, 0.31, 0.91],
    [0.1, 0.36, 0.64, 0.2, 0.56, 0.82],
    [0.13, 0.26, 0.38, 0.51, 0.63, 0.76, 0.88, 0.07]
]

week4_inputs = [
    [0.427778, 0.572222],
    [0.734259, 0.265741],
    [0.380952, 0.457143, 0.685714],
    [0.245238, 0.754762, 0.359524, 0.735714],
    [0.070833, 0.51, 0.929167, 0.26],
    [0.173704, 0.478148, 0.795926, 0.277407, 0.89963],
    [0.103636, 0.367879, 0.645455, 0.200606, 0.561818, 0.826061],
    [0.135, 0.261667, 0.385, 0.511667, 0.635, 0.761667, 0.885, 0.0775]
]

week5_inputs = [
    [0.42, 0.58],
    [0.63, 0.37],
    [0.31, 0.52, 0.74],
    [0.27, 0.61, 0.44, 0.83],
    [0.18, 0.57, 0.89, 0.33],
    [0.22, 0.48, 0.79, 0.29, 0.91],
    [0.14, 0.39, 0.66, 0.23, 0.58, 0.84],
    [0.16, 0.28, 0.41, 0.53, 0.66, 0.78, 0.9, 0.11]
]

# ---------------------------------------------------------
# HARD-CODED WEEKLY OUTPUTS
# ---------------------------------------------------------

week1_outputs = [
    5.724814756293084e-48,
    0.1668524240531267,
    -0.03513300955356602,
    -16.181877221146838,
    94.62121154785156,
    -1.7666841443101928,
    1.062660972512584,
    8.668221875
]

week2_outputs = [
    2.6752879910742468e-09,
    0.43801389905941557,
    -0.06505619615899967,
    -15.300482957509526,
    73.84870222928335,
    -1.721714026598003,
    0.8380519063103001,
    8.528175
]

week3_outputs = [
    1.5542034905615116e-13,
    0.41155118475915997,
    -0.038979703371155035,
    -11.860925282389413,
    85.55341253928546,
    -1.8218346425529346,
    1.0054527181773487,
    8.63406
]

week4_outputs = [
    1.4138862577600168e-16,
    0.3110015499941023,
    -0.11851799494062873,
    -14.10998365700517,
    84.26606790996709,
    -1.8336621537886206,
    0.9691878328431802,
    8.611369718333
]

week5_outputs = [
    3.4004559045423516e-18,
    0.3658481167000386,
    -0.11876831843175635,
    -13.16435158471511,
    54.36741321156047,
    -1.7568980194796602,
    0.8328963722243514,
    8.50409
]

# ---------------------------------------------------------
# ML MODELS
# ---------------------------------------------------------

def get_models():
    return [
        RandomForestRegressor(n_estimators=200),
        GradientBoostingRegressor(),
        SVR(kernel='rbf', C=10, gamma='scale'),
        MLPRegressor(hidden_layer_sizes=(50, 50), max_iter=500)
    ]

# ---------------------------------------------------------
# GENERATE WEEK 6 INPUTS USING MODEL CONSENSUS
# ---------------------------------------------------------

def generate_week6_inputs():
    week6 = []

    all_inputs = [week1_inputs, week2_inputs, week3_inputs, week4_inputs, week5_inputs]
    all_outputs = [week1_outputs, week2_outputs, week3_outputs, week4_outputs, week5_outputs]

    for f in range(8):
        X = np.array([w[f] for w in all_inputs], dtype=object)
        X = np.array([np.array(x, dtype=float) for x in X])
        y = np.array([w[f] for w in all_outputs], dtype=float)

        models = get_models()
        for m in models:
            m.fit(X, y)

        base = np.array(week5_inputs[f])
        candidates = [base + np.random.uniform(-0.02, 0.02, size=len(base)) for _ in range(200)]
        candidates = np.clip(candidates, 0, 1)

        best_point = None
        best_var = float("inf")

        for c in candidates:
            preds = np.array([m.predict([c])[0] for m in models])
            disagreement = np.var(preds)

            if disagreement < best_var:
                best_var = disagreement
                best_point = c

        week6.append(best_point.tolist())

    return week6

# ---------------------------------------------------------
# PREDICT WEEK 6 OUTPUTS
# ---------------------------------------------------------

def predict_week6_outputs(week6_inputs):
    outputs = []

    all_inputs = [week1_inputs, week2_inputs, week3_inputs, week4_inputs, week5_inputs]
    all_outputs = [week1_outputs, week2_outputs, week3_outputs, week4_outputs, week5_outputs]

    for f in range(8):
        X = np.array([w[f] for w in all_inputs], dtype=object)
        X = np.array([np.array(x, dtype=float) for x in X])
        y = np.array([w[f] for w in all_outputs], dtype=float)

        models = get_models()
        best_model = None
        best_var = float("inf")

        # pick model with lowest disagreement
        for m in models:
            m.fit(X, y)
            preds = m.predict(X)
            var = np.var(preds)
            if var < best_var:
                best_var = var
                best_model = m

        pred = best_model.predict([week6_inputs[f]])[0]
        outputs.append(pred)

    return outputs

# ---------------------------------------------------------
# RUN EVERYTHING
# ---------------------------------------------------------

week6_inputs = generate_week6_inputs()
week6_outputs = predict_week6_outputs(week6_inputs)

print("=== WEEK 6 INPUTS ===")
for i, w in enumerate(week6_inputs, start=1):
    print(f"Function {i}: {w}")

print("\n=== WEEK 6 OUTPUTS ===")
for i, o in enumerate(week6_outputs, start=1):
    print(f"Function {i}: {o}")
```
